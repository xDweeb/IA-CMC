```python
#importation des bibliotheque
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC, SVR
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#chargement de la base de donnees
data=pd.read_csv("bill_authentication.csv")
```

```python
#Afficher l'en tete de la DB et sa dimension
print(data.head(),\
      "\nDimension:",data.shape)

```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
x_train ,x_test ,y_train ,y_test = train_test_split(x,y,test_size=1.0/3)
```

```python
classifier = SVC(kernel = "linear", random_state = 0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
sb.pairplot(data,hue="Class",palette="Set1")
```

```python
#matrice de confusion et visualisation
cm=confusion_matrix(y_test,y_pred)
sb.heatmap(cm,annot=True)
```

```python
# Rapport de classification
print(classification_report(y_test,y_pred))
```

```python
#1 Chargement de bd "IRIS"
irisdata=pd.read_csv("iris.csv")
```

```python
#2 l'en tete de base de donnees
irisdata.head()
```

```python
# 3 interpretation des resultat

sb.pairplot(irisdata,hue="variety",palette="Set2")
```

```python
# 4
Q=irisdata.iloc[:,:-1].values
Z=irisdata.iloc[:,-1].values
Q_train,Q_test,Z_train,Z_test=train_test_split(Q,Z,test_size=1.0/3)
```

#5
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report

```python
# 6
clfr_linear= SVC(kernel="linear",random_state=1)
```

```python
clfr_linear.fit(Q_train,Z_train)
```

```python
prediction_L=clfr_linear.predict(Q_test)
print(prediction_L)
```

```python
# 7 Calcul de precision
clfr_linear.score(Q_test,Z_test)
```

```python
# 8 Matrice de confusion
conf_matrx=confusion_matrix(Z_test,prediction_L)
```

```python
# visualisation dela matrice de confusion
sb.heatmap(conf_matrx,annot=True)
```

```python
#9 Rapport de classification
rc=classification_report(Z_test,prediction_L)
print(rc)
```

```python
# 10 - 11
clfr_rbf=SVC(kernel="rbf",gamma=0.001)
clfr_rbf.fit(Q_train,Z_train)
```

```python
prediction_rbf=clfr_rbf.predict(Q_test)
```

```python
prediction_rbf
```

```python
clfr_rbf.score(Q_test,Z_test)
```

```python
#12 Matrice de confusion
conf_matrx2=confusion_matrix(Z_test,prediction_rbf)

#visualisation dela matrice de confusion
sb.heatmap(conf_matrx2,annot=True)
```

```python
#13
print(classification_report(Z_test,prediction_rbf))
```

```python
#14

from sklearn import model_selection
C_range = np.logspace(-2, 3, 6)
gamma_range = np.logspace(-2, 1, 4)
param_grid = {'C': C_range, 'gamma': gamma_range}
score = 'accuracy'
grid = model_selection.GridSearchCV(SVC(kernel='rbf'),
param_grid, cv=5)
grid.fit(Q_train, Z_train)
print("Les paramÃ¨ters optimaux sont {} avec un score de {:.2f}".format(grid.best_params_, grid.best_score_))
```

```python
#15

from sklearn import metrics
kmatrix = metrics.pairwise.rbf_kernel(Q_train, gamma=0.1)
plt.pcolor(kmatrix, cmap=plt.cm.PuRd)
plt.colorbar()
plt.gca().invert_yaxis()
plt.gca().xaxis.tick_top()
plt.show()
```
