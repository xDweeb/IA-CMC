```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
from sklearn.ensemble import RandomForestClassifier
```

```python
oli=fetch_olivetti_faces()
```

```python
x , y = oli.data , oli.target
```

```python
mask = y < 5
x=x[mask]
y=y[mask]
```

```python
foret=RandomForestClassifier(n_estimators=750)
foret.fit(x,y)
```

```python
importances = foret.feature_importances_
img_shape = oli.images[0].shape
imp_reshaped = importances.reshape(img_shape)
```

```python
import matplotlib.pyplot as plt
plt.matshow(imp_reshaped, cmap=plt.cm.hot)
plt.title("L'importance des pixels en utilisant la valeur des impuretés")
plt.colorbar()
plt.show()
```
